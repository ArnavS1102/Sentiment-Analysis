```python
df1=df1.sort_values(by=['date'])
```

```python
df1=df1.reset_index()
df1.drop("index",axis=1,inplace=True)
```

## Cleaning data

```python
new_dfs=pd.DataFrame({"tweet":[]})
for i in range(df1.shape[0]):   
    x=re.sub('#\S+',' ',str(df1.loc[i,"tweet"]))
    x=re.sub('@\S+',' ',x)
    x=re.sub('\S+@\S+',' ',x)
    x=re.sub('https\S+',' ',x)
    x=re.sub('[^A-Za-z0-9]',' ',x)
    x=re.sub('\d+',' ',x)
    li=[]
    for l in x.lower().split():
        if l.isalpha()==True:
            li.append(l)
    str2=""
    for l in li:
        if l not in stop_li:
                str2+=l+" "
    str_temp = word_tokenize(str2.lower())
    str3=""
    for k in str_temp:
        str3+=lemmatizer.lemmatize(k)+" "
    new_dfs.at[i,"tweet"]=str3.strip()

```

## Using pre trained word embeddings for an embedding layer of the neural network

```python
from sklearn.utils import shuffle
from nltk.tokenize import word_tokenize
import itertools
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import codecs
from tqdm import tqdm
import numpy as np
li_train1=[]
for i in new_dfs["tweet"]:
    words=word_tokenize(i)
    li_train1.append(words)
li_train = list(itertools.chain(*li_train1))
li_all=li_train
li_uniquetrain=set(li_train)
li_unique=set(li_train)
tokenizer = Tokenizer(num_words=len(li_unique), lower=True, char_level=False)
tokenizer.fit_on_texts(li_all)
word_seq_train = tokenizer.texts_to_sequences(li_train1)
word_index = tokenizer.word_index
li_all1=li_train1
max=0
for i in li_all1:
    if len(i)>max:
        max=len(i)
train=pad_sequences(word_seq_train, maxlen=100)
embeddings_index = {}
f = codecs.open('glove.twitter.27B.200d.txt', encoding='utf-8')
for line in tqdm(f):
    values = line.rstrip().rsplit(' ')
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
words_not_found = []
nb_words = min(li_unique.__len__(), len(word_index)+1)
embedding_matrix = np.zeros((nb_words, 200))
for word, i in word_index.items():
    if i >= nb_words:
        continue
    embedding_vector = embeddings_index.get(word)
    if (embedding_vector is not None) and len(embedding_vector) > 0:
        embedding_matrix[i] = embedding_vector
    else:
        words_not_found.append(word)
```

## Loading a trained neural network for sentiment classification

```python
from keras.models import load_model
model = load_model('lstm_sentiment1.h5')
predictions = model.predict(train)
```

## Probability distribution of each sentiment class i.e positive and negative is calculated by the final layer of the neural network.

```python
predictions
```

## This probability distribution is converted to a single sentiment.

```python
final=list([list(x) for x in predictions])
y_prediction=[]
for x in final:
    if x[0]>x[1]:
        y_prediction.append(0)
    elif x[0]<x[1]:
        y_prediction.append(1)
    else:
        y_prediction.append(2)
```

```python
y_prediction
```

```python
new_dfr=pd.DataFrame({"date":[],"tweet":[],"sentiment":[]})
new_dfr["date"]=df1["date"]
new_dfr["tweet"]=df1["tweet"]
new_dfr["sentiment"]=y_prediction
```

```python
for i in range(new_dfr.shape[0]):
    new_dfr.loc[i,"date"]=new_dfr.loc[i,"date"][:10]
```

```python
li_date=list(set(new_dfr["date"].values))
```

```python
li_date
```

```python
li_sentidate=[]
for i in li_date:
    pos=0
    neg=0
    for j in range(new_dfr.shape[0]):
        if new_dfr.loc[j,"sentiment"]==1 and new_dfr.loc[j,"date"]==i:
            pos+=1
        elif new_dfr.loc[j,"sentiment"]==0 and new_dfr.loc[j,"date"]==i:
            neg+=1
    li_sentidate.append([pos,neg])
    
```

```python
li_sentidate
```

```python
df_new=pd.DataFrame({"date":[],"positive":[],"negative":[]})
```

```python
df_new["date"]=li_date
df_new["positive"]=[(x[0]/(x[0]+x[1]))*100 for x in li_sentidate]
df_new["negative"]=[(x[1]/(x[0]+x[1]))*100 for x in li_sentidate]
```
